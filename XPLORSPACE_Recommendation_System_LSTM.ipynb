```python
#Creating two corpuses to hold the data about the abstract and keywords 
```

```python
x=combined.iloc[:,[0]].values.tolist()   
x1=combined.iloc[:,[1]].values.tolist()
len(x)
corpus=[]
corpus1=[]
k=[]
for i in range(0,len(x)):
  review=re.sub('[^a-zA-Z]','',str(x[i]))
  review=review.lower()
  review=review.split()
  review = [Lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
  review=''.join(review)
  corpus.append(review)
```

```python
for i in range(0,len(x1)):
  review=re.sub('[^a-zA-Z]','',str(x1[i]))  #removing spaces and other characters
  review=review.lower()
  review=review.split()
  review = [Lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')]  #applying lemmatization
  review=''.join(review)
  corpus1.append(review)
```

```python
corpus1=pd.DataFrame(corpus1)
corpus=pd.DataFrame(corpus)

```

```python
corpus_final=corpus1+corpus   #Adding the keywords with the existing text 
# The pattern would be keyword+abstract
```

```python
corpus_final=corpus_final.iloc[:,0].values.tolist()
```

```python
corpus_final
```

```python
y=combined.iloc[:,[2]].values
y=LE.fit_transform(y)
```

```python
len(corpus_final)
sentence_length=80 # assumed a longer sentence length of 2000 words
OHR=[one_hot(k,4000) for k in corpus_final] #generating one hot encodings
embeddings=pad_sequences(OHR,padding='pre',maxlen=sentence_length)  #padding the sequences so that all of them are of equal length

x_train,x_test,y_train,y_test=train_test_split(embeddings,y,train_size=0.8,shuffle=True) #train_test_split
```

LSTM Model
This model severly overfits. Because of limited data.

```python
from keras.layers import Dense
from keras.models import Sequential, save_model,load_model
from keras.layers import LSTM,Bidirectional
from keras.layers import Dense,Embedding,LeakyReLU
from keras.models import Sequential
from keras.layers import LSTM,Bidirectional
import pickle
model=Sequential()
model.add(Embedding(4000,4,input_length=sentence_length)) #Embedding Layer
model.add(BatchNormalization())
model.add(Bidirectional(LSTM(1,activation='tanh',return_sequences=True)))
#model.add(LeakyReLU(alpha=0.3))
model.add(LSTM(2,activation='tanh',return_sequences=False))
model.add(Dense(4,activation='softmax'))
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=20,batch_size=15)
```
